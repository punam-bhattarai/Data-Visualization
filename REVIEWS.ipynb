9. Load the Google Playstore dataset and examine its dimensions. How many rows and columns does the dataset have?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
df.info()
df
```

```python
df['Reviews'].str.isnumeric().sum()
```

```python
df[~df['Reviews'].str.isnumeric()]
```

```python
df['Reviews'] = df['Reviews'].str.replace('3.0M', '3000000')
```

```python
df['Reviews']= df['Reviews'].astype('int')
```

```python
df.drop(index = 10472, inplace = True)
```

```python
df['Size'].unique()
```

```python
def convert_size(size):
  if 'M' in size:
    return int(float(size.replace('M', ''))*1000)
  elif 'k' in size:
    return float(size.replace('k', ''))
  elif 'Varies with device' in size:
    return np.nan
  else:
    return np.nan
```

```python
df['Size']=df['Size'].apply(convert_size)
df.head()
```

```python
df['Price'].unique()
```

```python
df['Installs'].unique()
```

```python
cols= ['Installs','Price']
chars_to_remove = ['+', ',', '$']
for i in cols:
  for j in chars_to_remove:
    df[i]=df[i].str.replace(j, '')
```

```python
df['Installs']=df['Installs'].astype('int')
df['Price']=df['Price'].astype('float')
```

```python
df['Last Updated']=pd.to_datetime(df['Last Updated'])
df['Day']=df['Last Updated'].dt.day
df['Month']=df['Last Updated'].dt.month
df['Year']=df['Last Updated'].dt.year
df.info()
```

```python
df.drop_duplicates(subset= ['App'], keep='first', inplace= True)
df
```

14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.


```python

installs_frequency_table = df.groupby('App')['Installs'].sum().reset_index()
top_10_app = installs_frequency_table.sort_values(by='Installs', ascending=False).head(10)
print("Frequency Table :\n", installs_frequency_table)
print('---------------------------------------------------------------------')
print("Top 10 most popular apps with the highest number of installs are :\n", top_10_app)
plt.figure(figsize=(10,5))
plt.barh(top_10_app['App'],top_10_app['Installs'])
plt.xlabel('Number of Installs')
plt.ylabel('App')
plt.title('Top 10 Most Popular Apps by Number of Installs')
```

15. A company wants to launch a new app on the Google Playstore and has asked you to analyze the Google Playstore dataset to identify the most popular app categories. How would you approach this task, and what features would you analyze to make recommendations to the company?


To figure out which types of apps are popular on the Google Play Store, I'd first check out how many apps there are in different categories. Then, I'd see how many times apps in each category have been downloaded and what their average ratings are. Based on this info, I'd suggest focusing on categories where apps are really popular and liked by users. To show this info visually, I'd make a bar chart that highlights the top 10 most popular apps by how many times they've been downloaded. This chart would make it easy to see which app categories are the most popular, helping the company make smart decisions.


```python
popular_app = df.groupby('Category').agg({'Installs': 'sum', 'Rating': 'mean'}).reset_index()

top_10_app = popular_app.sort_values(by=['Installs', 'Rating'], ascending=False).head(10)

plt.figure(figsize=(20, 7))
plt.bar(top_10_app['Category'], top_10_app['Installs'])
plt.xlabel('App Category')
plt.ylabel('Total Installs')
plt.title('Top 10 Most Popular Apps by Installs')
plt.show()

```

16. A mobile app development company wants to analyze the Google Playstore dataset to identify the most successful app developers. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?


We're analyzing different app developers based on three things: how many times their apps have been installed, the average ratings of their apps, and the total number of reviews their apps have received. We group the data by developer and then sort them by the total number of installs to find the top 10 most successful developers. Finally, we visualize this information using a horizontal bar chart, making it easy to see which developers have the highest number of installs.

```python
developer_stats = df.groupby('Category').agg({
    'Installs': 'sum',
    'Rating': 'mean',
    'Reviews': 'sum'
}).reset_index()

developer_stats = developer_stats.sort_values(by='Installs', ascending=False)
top_10_developers = developer_stats.head(10)

```

```python
print(top_10_developers.shape)
```

```python
sns.barplot(x='Installs', y='Category', data=top_10_developers, color='teal')
plt.xlabel('Total Installs')
plt.ylabel('App Developer')
plt.title('Top 10 App Developers by Total Installs')
plt.show()
```

17. A marketing research firm wants to analyze the Google Playstore dataset to identify the best time to launch a new app. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

The solution looks at different parts of app launches to find patterns and trends, helping decide the best time to launch a new app on Google Play. It checks when apps are released throughout the year to see if there are any trends or times when more apps are released. It also looks at how many times apps are installed, their ratings, and reviews to see if there's any difference depending on when they were released. The data is split up by the month and year of release, and then averages for installs, ratings, and reviews are calculated for each group. Then, there are two graphs made to show how these things change each month and each year.

```python
monthly_stats = df.groupby('Month').agg({
    'Installs': 'mean',
    'Rating': 'mean',
    'Reviews': 'mean'
}).reset_index()


yearly_stats = df.groupby('Year').agg({
    'Installs': 'mean',
    'Rating': 'mean',
    'Reviews': 'mean'
}).reset_index()

plt.figure(figsize=(10, 5))
plt.plot(monthly_stats['Month'], monthly_stats['Installs'], label='Average Installs')
plt.plot(monthly_stats['Month'], monthly_stats['Rating'], label='Average Rating')
plt.plot(monthly_stats['Month'], monthly_stats['Reviews'], label='Average Reviews')
plt.xlabel('Month')
plt.ylabel('Metrics')
plt.title('Monthly Trends for App Launches')
plt.legend()
plt.grid(True)
plt.xticks(range(1, 13))
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(yearly_stats['Year'], yearly_stats['Installs'], label='Average Installs')
plt.plot(yearly_stats['Year'], yearly_stats['Rating'], label='Average Rating')
plt.plot(yearly_stats['Year'], yearly_stats['Reviews'], label='Average Reviews')
plt.xlabel('Year')
plt.ylabel('Metrics')
plt.title('Yearly Trends for App Launches')
plt.legend()
plt.grid(True)

```
